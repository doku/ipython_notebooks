```python
def kmer(k, text, t):
    result = {}
    for i in xrange(len(text)-k):
        if text[i:i+k] in result:
            result[text[i:i+k]] += 1
        else:
            result[text[i:i+k]] = 1
        
    return [j for j,k in result.iteritems() if k == t]
```
